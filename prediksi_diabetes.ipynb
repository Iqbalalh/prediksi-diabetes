```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# load data dari github

dataset = pd.read_csv("https://raw.githubusercontent.com/Iqbalalh/prediksi-diabetes/main/diabetes.csv")
dataset
```

Preprocessing

```python
# drop skin thickness untuk efisiensi
df = dataset.drop(['SkinThickness'], axis = 1)
```

```python
df
```

Train Test Split

```python
# split train dan test data
df_train, df_test = train_test_split(df, test_size=0.4, random_state=82)
```

```python
df_train
```

```python
# memisahkan fitur dengan variabel yg dijadikan target

X_train = df_train.drop('Outcome', axis=1)
y_train = df_train['Outcome']
X_test = df_test.drop('Outcome', axis=1)
y_test = df_test['Outcome']
```

```python
# pre processing dengan scaling dataset menggunakan minmaxscaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# seleksi fitur menggunakan chi squared test

selector = SelectKBest(chi2, k=7)
X_train_new = selector.fit_transform(X_train_scaled, y_train)
X_test_new = selector.transform(X_test_scaled)
```

```python
# fitur yang akan dipakai untuk training model

X_train.columns[selector.get_support()]
```

Model menggunakan Logistic Regression

```python
# model

prediksi_diabetes = LogisticRegression(penalty='l2', random_state=45645645, solver='saga')
prediksi_diabetes.fit(X_train_new, y_train)
```

Evaluasi

```python
# accuracy

accuracy = prediksi_diabetes.score(X_test_new, y_test)
accuracy
```
